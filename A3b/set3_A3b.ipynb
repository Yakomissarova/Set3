```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('intro_comparative_analysis.csv')
df
```

```python
# График 1: Обычный Quick Sort на всех типах входных данных
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(df['Size'], df['Random_Standard'], label='Случайные массивы', linewidth=2, color='blue')
plt.plot(df['Size'], df['Reverse_Standard'], label='Обратно отсортированные', linewidth=2, color='red')
plt.plot(df['Size'], df['Almost_Standard'], label='Почти отсортированные', linewidth=2, color='green')
plt.xlabel('Размер массива')
plt.ylabel('Время (микросекунды)')
plt.title('Standard Quick Sort\nВсе типы входных данных')
plt.legend()
plt.grid(True, alpha=0.3)

# График 2: Introsort на всех типах входных данных
plt.subplot(1, 2, 2)
plt.plot(df['Size'], df['Random_Intro'], label='Случайные массивы', linewidth=2, color='blue')
plt.plot(df['Size'], df['Reverse_Intro'], label='Обратно отсортированные', linewidth=2, color='red')
plt.plot(df['Size'], df['Almost_Intro'], label='Почти отсортированные', linewidth=2, color='green')
plt.xlabel('Размер массива')
plt.ylabel('Время (микросекунды)')
plt.title('Introsort\nВсе типы входных данных')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

Быстрее всего оба алгоритма сортируют почти отсортированные массивы, в то время как случайные массивы сортируются дольше всего.

```python
# График сравнения Quick Sort vs Introsort
plt.figure(figsize=(15, 5))

# Случайные массивы
plt.subplot(1, 3, 1)
plt.plot(df['Size'], df['Random_Standard'], label='Standard Quick Sort', linewidth=2, color='blue')
plt.plot(df['Size'], df['Random_Intro'], label='Introsort', linewidth=2, color='green')
plt.xlabel('Размер массива')
plt.ylabel('Время (микросекунды)')
plt.title('Случайные массивы')
plt.legend()
plt.grid(True, alpha=0.3)

# Обратно отсортированные массивы
plt.subplot(1, 3, 2)
plt.plot(df['Size'], df['Reverse_Standard'], label='Standard Quick Sort', linewidth=2, color='blue')
plt.plot(df['Size'], df['Reverse_Intro'], label='Introsort', linewidth=2, color='green')
plt.xlabel('Размер массива')
plt.ylabel('Время (микросекунды)')
plt.title('Обратно отсортированные массивы')
plt.legend()
plt.grid(True, alpha=0.3)

# Почти отсортированные массивы
plt.subplot(1, 3, 3)
plt.plot(df['Size'], df['Almost_Standard'], label='Standard Quick Sort', linewidth=2, color='blue')
plt.plot(df['Size'], df['Almost_Intro'], label='Introsort', linewidth=2, color='green')
plt.xlabel('<UNK> <UNK>')
plt.ylabel('<UNK> (<UNK>)')
plt.xlabel('Размер массива')
plt.ylabel('Время (микросекунды)')
plt.title('Почти отсортированные массивы')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

Вывод:
По графикам видно, что Introsort на больших значениях массива, сортирует его гораздо быстрее чем просто Quick Sort. Также с увеличением размера массива, время затраченное на Introsort увеличивается медленнее чем время потраченное на Quick Sort.
Следовательно, для максимального укорения сортировки необходимо использовать Introsort как лучший по времени вариант.
